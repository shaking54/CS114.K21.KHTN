# Load dữ liệu

```python
sales = turicreate.SFrame('/content/home_data.sframe/m_1ce96d9d245ca490.frame_idx')
```

```python
sales
```

# Biểu diễn từng đặc tính của căn nhà để đánh giá dữ liệu

```python
sales.show()
```

```python
turicreate.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

# Xây dựng một regression model đơn giản để dự đoán giá nhà từ diện tích

```python
training_set, test_set = sales.random_split(.8,seed=0)
```

## Huấn luyện regression model

```python
sqft_model = turicreate.linear_regression.create(training_set,target='price',features=['sqft_living'])
```

# đánh giá model

```python
print (test_set['price'].mean())
```

```python
print (sqft_model.evaluate(test_set))
```

# Tìm hiểu sâu thêm về model vừa được xây dựng

```python
sqft_model.coefficients
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_set['sqft_living'],test_set['price'],'.',
        test_set['sqft_living'],sqft_model.predict(test_set),'-')
```
